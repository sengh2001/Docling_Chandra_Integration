```python
from docling_core.types.doc.document import DoclingDocument
from src.matching import align_english_hindi_blocks
```

```python
hindi_json_path = "outputs_split/Class_6-Science-Hindi/Chapter_3/merged.json"
english_json_path = "outputs_split/Class_6-Science-English/Chapter_3/merged.json"

hindi_doc = DoclingDocument.load_from_json(hindi_json_path)
english_doc = DoclingDocument.load_from_json(english_json_path) 
```

```python
hindi_blocks = {}
english_blocks = {}

for text_item in hindi_doc.texts:
    hindi_blocks[text_item.self_ref] = text_item.text

for text_item in english_doc.texts:
    english_blocks[text_item.self_ref] = text_item.text
```

```python
eng2hin, hin2eng = align_english_hindi_blocks(
    english_blocks,
    hindi_blocks,
)
```

```python
for eng_id, hin_id in eng2hin.items():
    print("English:", english_blocks[eng_id])
    print("Hindi:", hindi_blocks[hin_id])
    print("\n")
```

```python
print("\nHindi -> English")
print(hin2eng)
```
