```python
from src.merge_document_scans import merge_documents
from src.bbox_draw import draw_bboxes_on_pdf
from docling_core.types.doc.document import DoclingDocument#, TableItem
```

```python
# books = ["Class_6-Science-Tamil"]
books = ["Class_6-Science-English", "Class_6-Science-Hindi", "Class_6-Science-Tamil"]
chapters = list(range(1,13))

for book in books:
    for chapter in chapters:
        chapter_id = f"{book}/Chapter_{chapter}"
        chapter_jsons_directory = "./outputs_split/" + chapter_id
        pdf_path = "./books/" + chapter_id + ".pdf"

        full_scan_path = chapter_jsons_directory + "/parsed_original.json"
        text_scan_path = chapter_jsons_directory + "/parsed_text.json"
        merged_json_path = chapter_jsons_directory + "/merged.json"
        merged_annotated_pdf_path = chapter_jsons_directory + "/merged_annotated.pdf"
        original_annotated_pdf_path = (
            chapter_jsons_directory + "/original_annotated.pdf"
        )
        text_annotated_pdf_path = chapter_jsons_directory + "/text_annotated.pdf"

        full_doc = DoclingDocument.load_from_json(full_scan_path)
        text_doc = DoclingDocument.load_from_json(text_scan_path)

        merged_doc = merge_documents(full_doc, text_doc, verbose=False)

        merged_doc.save_as_json(merged_json_path)

        draw_bboxes_on_pdf(merged_doc, pdf_path, merged_annotated_pdf_path)
        # draw_bboxes_on_pdf(full_doc, pdf_path, original_annotated_pdf_path)
        # draw_bboxes_on_pdf(text_doc, pdf_path, text_annotated_pdf_path)
```
