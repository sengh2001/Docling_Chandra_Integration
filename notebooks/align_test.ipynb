```python
from docling_core.types.doc import DocItemLabel, DoclingDocument

from src.merging.merge_helpers import _find_best_match, _group_by_page, _get_bbox_for_item
```

```python
eng_doc = DoclingDocument.load_from_json("./outputs_split/Class_6-Science-English/Chapter_2/merged.json")
tam_doc = DoclingDocument.load_from_json("./outputs_split/Class_6-Science-Hindi/Chapter_2/merged.json")
```

```python
target_labels = {DocItemLabel.CAPTION, DocItemLabel.TEXT, DocItemLabel.SECTION_HEADER}
eng_pages = _group_by_page(eng_doc, target_labels)
tam_pages = _group_by_page(tam_doc, target_labels)
```

```python
page = 1
common_pages = set(eng_pages.keys()) & set(tam_pages.keys())
for page in common_pages:
    
    eng_page = eng_pages[page]
    tam_page = tam_pages[page]


    for e in eng_page:
        e_bbox = _get_bbox_for_item(e)
        
        for t in tam_page:
            t_bbox = _get_bbox_for_item(t)
            
            if e_bbox.overlaps(t_bbox):
                iou = e_bbox.intersection_over_union(t_bbox)
                if iou > 0.3:
                    print("IoU : ", iou)
                    print(e.text)
                    print(t.text)
```
